Reduction operation


```python
e= torch.randint(size=(2,3), low=0, high=10,dtype=torch.float32)
e
```

```python
# sum
a=torch.sum(e)

# sum along columns
b= torch.sum(e, dim=0)

# sum along rows

c= torch.sum(e, dim=1)

print(a)
print(b)
print(c)
```

```python
# mean
a=torch.mean(e)

# mean along columns
b= torch.mean(e, dim=0)

# mena along rows
c= torch.mean(e, dim=1)

print(a)
print(b)
print(c)
```

```python
# median

a= torch.median(e)

#median along columns
b= torch.median(e, dim =0)

# median along rows
c= torch.median(e,dim=1)

print(a)
print(b)
print(c)
```

```python
# max and min
a=torch.max(e)
b= torch.min(e)

# max and min along columns and rows
c=torch.max(e,dim=0)
d=torch.min(e,dim=1)

print(a)
print(b)
print(c)
print(d)
```

```python
# product
print(torch.prod(e))
print(torch.prod(e,dim=0))
print(torch.prod(e,dim=1))
```

```python
# standard deviation and variance

#std..cols,rows

print(torch.std(e))
print(torch.std(e,dim=0))
print(torch.std(e,dim=1))


# variane..cols, rows.

print(torch.var(e))
print(torch.var(e,dim=0))
print(torch.var(e,dim=1))
```

```python
# argmax ( index position of the largest element)
print(torch.argmax(e))
print(torch.argmax(e,dim=0))
print(torch.argmax(e,dim=1))
```

```python
#argmin (index position of the smallest element)
print(torch.argmin(e))
print(torch.argmin(e,dim=0))
print(torch.argmin(e,dim=1))
```

```python
# why we are using cpu and gpu .... when the ans and wokring is almost same right ??

# here's the answer:

import torch
import time

# check device
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# matrix size
size = 1000

# create CPU matrices
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# CPU timing
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)
cpu_time = time.time() - start_time
print(f"Time on CPU: {cpu_time:.4f} seconds")

# move matrices to GPU (do this before timing)
matrix_gpu1 = matrix_cpu1.to(device)
matrix_gpu2 = matrix_cpu2.to(device)

# GPU timing
start_time = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)
torch.cuda.synchronize()  # wait for GPU to finish
gpu_time = time.time() - start_time
print(f"Time on GPU: {gpu_time:.4f} seconds")

# compare results (should be almost equal)
print("Difference:", torch.norm(result_cpu - result_gpu.cpu()).item())


```
